```python
# Dropping the date column
df = df.drop(['Date'],axis=1)
df
```

```python
# For column evaporation, replace all missing values with average of the column
df['Evaporation'].fillna(df['Evaporation'].mean() , inplace=True)
df
```

```python
df.dropna(axis='columns' , how='any')
```

```python
#Making a user defined function where all details of country entered by the user are displayed
def ShowData(country):
    filt = (df['Location'] == country)
    print(df.loc[filt])
country = input("Enter country whose details you want to filter")
ShowData(country)
```
